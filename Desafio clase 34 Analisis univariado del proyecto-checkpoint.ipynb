### Analisis univarido de precio de las propiedades (SalePrice)

```python
df.SalePrice.describe()
```

```python
violinplot(y = df.SalePrice, data=df , orient="v")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,dpi=120,figsize = (8,4))

plot0=sns.distplot(x = df[df['SalePrice']!=0]['SalePrice'],ax=axes[0],color='green')
axes[0].yaxis.set_major_formatter(FormatStrFormatter('%f'))
axes[0].set_title('Distribution of Property price',fontdict={'fontsize':8})
axes[0].set_xlabel('Price',fontdict={'fontsize':7})
axes[0].set_ylabel('Property count',fontdict={'fontsize':7})

plot1=sns.boxplot(y = df[df['SalePrice']!=0]['SalePrice'],ax=axes[1],orient='v')
axes[1].set_title('Numerical Summary',fontdict={'fontsize':8})
axes[1].set_xlabel('SalePrice',fontdict={'fontsize':7})
axes[1].set_ylabel(r'Property price',fontdict={'fontsize':7})
plt.tight_layout()
```
